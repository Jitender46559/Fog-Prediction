```
# Importing important libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.naive_bayes import GaussianNB
from sklearn import model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```
from google.colab import files
uploaded = files.upload()
import io
data = pd.read_csv(io.BytesIO(uploaded['Iris.csv']))
```

```
data.head()
```

```
#Replace Species type with numbers
data.Species.replace(['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'], [1, 2, 3], inplace=True)
```

```
# Split-out validation dataset
split_data = data.values
train_data = split_data[:,1:5]
target = split_data[:,5]
train_feature, test_feature , train_label, test_label = train_test_split(train_data, target, 
                                                    test_size=0.3, 
                                                    random_state=7)
```

```
train_feature.shape, test_feature.shape, train_label.shape, test_label.shape
```

```
#Initialize Gaussian Naive Bayes
clf = GaussianNB()
```

```
# Test options and evaluation metric
scoring = 'accuracy'

#Fitting the training set
clf.fit(train_feature, train_label) 

#Predicting for the Test Set
pred_clf = clf.predict(test_feature)
```

```
#Prediction Probability
prob_pos_clf = clf.predict_proba(test_feature)[:, 1]
```

```
#Model Performance
#setting performance parameters
kfold = model_selection.KFold(n_splits=10, random_state=7)

#calling the cross validation function
cv_results = model_selection.cross_val_score(GaussianNB(), train_feature, train_label, cv=kfold, scoring=scoring)

#displaying the mean and standard deviation of the prediction
msg = "%s: %f (%f)" % ('NB accuracy', cv_results.mean(), cv_results.std())
print(msg)
```
